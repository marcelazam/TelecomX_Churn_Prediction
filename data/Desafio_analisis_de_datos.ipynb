```python
import matplotlib.pyplot as plt

# Datos de ingresos
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingresos_tienda_1, ingresos_tienda_2, ingresos_tienda_3, ingresos_tienda_4]

# Crear gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(tiendas, ingresos, color='skyblue')
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos Totales')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Mostrar gráfico
plt.show()
```

```python
# Datos de categorías más vendidas
categorias_tienda_1 = categorias_vendidas_tienda_1.index
ventas_tienda_1 = categorias_vendidas_tienda_1.values

# Crear gráfico de barras
plt.figure(figsize=(12, 6))
plt.bar(categorias_tienda_1, ventas_tienda_1, color='lightgreen')
plt.title('Categorías Más Vendidas - Tienda 1')
plt.xlabel('Categorías del Producto')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Mostrar gráfico
plt.show()
```
